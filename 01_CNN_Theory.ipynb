## **Stride**

It controls the boxes to overlap 

How many spaces to move the pulling box for each output

In CNN:

The input image shrinks.

Since the filters stay in the same size, they find increasingly large patterns 

Therefore, first, CNN learns small patterns and then bigger patterns.
This is why CNN learns hierarchical features.

Generally, the number of features maps increase at each layer.



# **Some guidlines for hyperparameters in CNN**



1.   Small filters relative to the images such as 3 * 3, 5 * 5 , and 7 * 7
2.   Repeat: conv-->pool-->conv-->pool-etc

3. Increase # feature maps such as 32-->64-->128--> 128
